## 📌 1. KISIM: Veri Yapıları ve Temel Mantık
**Amaç:** Tek bir müşterinin verilerini Python veri yapılarıyla temsil etmek ve basit kararlar verdirmek.

```python
# ========================
# GÖREV 1: Değişkenler ve Tipler
# ========================

ad_soyad = "Ahmet Yılmaz"          # String (Metin) Tipi — Müşteri adı ve soyadı
aylik_ucret = 650.0                 # Float (Ondalıklı Sayı) Tipi — Aylık ücret (TL)
sadakat_ayi = 30                    # Integer (Tam Sayı) Tipi — Kaç aydır müşteri
aktif_mi = True                     # Boolean (Mantıksal) Tipi — Hesap aktif mi?

# Değişken tiplerini ekrana yazdırma
print("=== MÜŞTERİ DEĞİŞKENLERİ ===")
print(f"Ad Soyad     : {ad_soyad}  → Tip: {type(ad_soyad).__name__}")
print(f"Aylık Ücret  : {aylik_ucret} TL → Tip: {type(aylik_ucret).__name__}")
print(f"Sadakat Ayı  : {sadakat_ayi} ay → Tip: {type(sadakat_ayi).__name__}")
print(f"Aktif Mi?    : {aktif_mi}  → Tip: {type(aktif_mi).__name__}")
```

```python
# ========================
# GÖREV 2: Liste ve Sözlük Kullanımı
# ========================

# Liste Kullanımı — Şirketin sunduğu 5 farklı hizmet
hizmetler_listesi = [
    "Fiber İnternet",      # Liste elemanı 1
    "Dijital TV",           # Liste elemanı 2
    "Mobil Hat",            # Liste elemanı 3
    "Bulut Depolama",       # Liste elemanı 4
    "Siber Güvenlik"        # Liste elemanı 5
]

# Sözlük (Dictionary) Kullanımı — Müşterinin tüm bilgileri
musteri = {
    "ad_soyad"    : ad_soyad,      # String değer — anahtar:değer yapısı
    "aylik_ucret" : aylik_ucret,   # Float değer
    "sadakat_ayi" : sadakat_ayi,   # Integer değer
    "aktif_mi"    : aktif_mi,      # Boolean değer
    "hizmetler"   : hizmetler_listesi  # Liste değer — iç içe yapı
}

# Çıktıları yazdırma
print("=== ŞİRKET HİZMETLERİ LİSTESİ ===")
for i, hizmet in enumerate(hizmetler_listesi, 1):  # Döngü ile listeleme
    print(f"  {i}. {hizmet}")

print("\n=== MÜŞTERİ SÖZLÜĞÜ ===")
for anahtar, deger in musteri.items():  # Sözlük döngüsü
    print(f"  {anahtar:15}: {deger}")
```

```python
# ========================
# GÖREV 3: Koşullu İfadeler (If-Else)
# ========================

# Koşullu İfade — VIP müşteri tespiti
if musteri["aylik_ucret"] > 500 or musteri["sadakat_ayi"] > 24:  # If koşulu
    print("✅ VIP Müşteri: İndirim Tanımla")  # Koşul doğruysa bu çalışır
else:                                          # Koşul yanlışsa
    print("📋 Standart Müşteri")              # Bu çalışır

# Koşul gerekçesini açıklayan detay
print(f"   → Aylık Ücret: {musteri['aylik_ucret']} TL (Eşik: 500 TL)")
print(f"   → Sadakat Ayı: {musteri['sadakat_ayi']} ay (Eşik: 24 ay)")
```

```python
# ========================
# GÖREV 4: String Operasyonları ve Random ID Üretimi
# ========================

import random  # Random kütüphanesini içe aktarma

# String Operasyonu — Büyük harfe çevirme
ad_buyuk = musteri["ad_soyad"].upper()  # String metodu: .upper()
print(f"Büyük Harf Ad : {ad_buyuk}")

# String Operasyonu — customerID formatında rastgele ID üretme
rastgele_sayi = random.randint(1000, 9999)           # Random kütüphanesi ile 4 haneli sayı üretimi
musteri_id = f"IST-2026-{rastgele_sayi}"             # f-string ile ID formatı oluşturma
musteri["musteri_id"] = musteri_id                   # Sözlüğe yeni anahtar ekleme

print(f"Müşteri ID    : {musteri_id}")
print(f"\n✅ Müşteri ID başarıyla oluşturuldu ve sözlüğe eklendi.")
```

### 💬 Kritik Soru Cevabı — 1. Kısım

> **"Müşteri bilgilerini neden bir 'Liste' yerine 'Sözlük' içinde saklamayı tercih ettik? Avantajı nedir?"**

**Cevap:**  
Listelerde verilere *indeks numarasıyla* erişilir (`musteri[0]`, `musteri[1]` gibi). Bu yaklaşım, hangi indeksin hangi bilgiyi tuttuğunu hatırlamayı zorlaştırır ve kodun okunabilirliğini düşürür.  

Sözlükte ise verilere *anlamlı bir anahtar (key)* ile erişilir: `musteri["ad_soyad"]`, `musteri["aylik_ucret"]`. Bu sayede:
- **Okunabilirlik** artar; kod kendini açıklar.
- **Bakım kolaylığı** sağlanır; yeni bilgi eklemek/çıkarmak diğer alanları etkilemez.
- **Erişim hızı** O(1)'dir; anahtar bilindiğinde doğrudan erişim mümkündür.
- **Veri bütünlüğü** korunur; yanlış indeks sıralamasından kaynaklanan hatalar önlenir.

Sonuç: Müşteri gibi *heterojen* (farklı tipte) ve *çok alanlı* verileri modellemek için sözlük çok daha uygun bir veri yapısıdır.

## 📌 2. KISIM: Fonksiyonlar, Döngüler ve Kütüphaneler
**Amaç:** Birden fazla müşteriyi yöneten ve analiz yapan bir sistem kurmak.

```python
# ========================
# GÖREV 1: Döngüler ve Listeler — 5 Müşterili Ana Liste
# ========================

import random    # Random kütüphanesi içe aktarma
import math      # Math kütüphanesi içe aktarma
import datetime  # Datetime kütüphanesi içe aktarma

# 5 farklı müşterinin sözlük verilerinden oluşan ana liste
musteriler_listesi = [                                              # Liste Kullanımı
    {
        "ad_soyad"    : "Ahmet Yılmaz",    # String — müşteri adı
        "aylik_ucret" : 650.0,              # Float — aylık ücret
        "sadakat_ayi" : 30,                 # Integer — sadakat süresi
        "aktif_mi"    : True,               # Boolean — hesap durumu
        "sehir"       : "İstanbul",         # String — şehir bilgisi
        "hizmetler"   : ["Fiber İnternet", "Dijital TV", "Mobil Hat"]  # Liste — hizmetler
    },
    {
        "ad_soyad"    : "Fatma Kaya",
        "aylik_ucret" : 299.0,
        "sadakat_ayi" : 6,
        "aktif_mi"    : True,
        "sehir"       : "Ankara",
        "hizmetler"   : ["Mobil Hat"]
    },
    {
        "ad_soyad"    : "Mehmet Demir",
        "aylik_ucret" : 520.5,
        "sadakat_ayi" : 18,
        "aktif_mi"    : False,              # Pasif müşteri — Churn riski!
        "sehir"       : "İzmir",
        "hizmetler"   : ["Fiber İnternet", "Bulut Depolama"]
    },
    {
        "ad_soyad"    : "Zeynep Arslan",
        "aylik_ucret" : 180.0,
        "sadakat_ayi" : 3,                  # Yeni müşteri — Churn riski!
        "aktif_mi"    : True,
        "sehir"       : "Bursa",
        "hizmetler"   : ["Mobil Hat"]
    },
    {
        "ad_soyad"    : "Can Öztürk",
        "aylik_ucret" : 890.0,
        "sadakat_ayi" : 48,
        "aktif_mi"    : True,
        "sehir"       : "İstanbul",
        "hizmetler"   : ["Fiber İnternet", "Dijital TV", "Mobil Hat", "Bulut Depolama", "Siber Güvenlik"]
    }
]

print(f"✅ Toplam {len(musteriler_listesi)} müşteri sisteme yüklendi.")  # len() fonksiyonu

# Müşterileri listeleme — For döngüsü
print("\n=== MÜŞTERİ LİSTESİ ===")
for i, m in enumerate(musteriler_listesi, 1):       # Döngü Kullanımı — enumerate ile indeks
    durum = "Aktif" if m["aktif_mi"] else "Pasif"   # Ternary koşullu ifade
    print(f"  {i}. {m['ad_soyad']:20} | {m['aylik_ucret']:>7.2f} TL | {m['sadakat_ayi']:>3} ay | {durum}")
```

```python
# ========================
# GÖREV 2: Fonksiyon Tanımlama — KDV Hesaplama
# ========================

def tutar_hesapla(aylik_ucret):          # Fonksiyon Tanımlama — def anahtar kelimesi
    """
    Verilen aylık ücrete %20 KDV ekleyerek KDV dahil tutarı döndürür.
    Parametre : aylik_ucret (float) — KDV hariç ücret
    Döndürür  : kdv_dahil_tutar (float) — KDV dahil tutar
    """
    kdv_orani = 0.20                             # Float — KDV oranı (%20)
    kdv_tutari = aylik_ucret * kdv_orani         # Aritmetik işlem — KDV miktarı
    kdv_dahil_tutar = aylik_ucret + kdv_tutari   # Aritmetik işlem — toplam tutar
    return kdv_dahil_tutar                       # Return — hesaplanan değeri geri döndür

# Fonksiyonu test etme
print("=== KDV HESAPLAMA FONKSİYONU TESTİ ===")
test_ucret = 500.0                               # Test değişkeni
sonuc = tutar_hesapla(test_ucret)               # Fonksiyon çağırma
print(f"KDV Hariç : {test_ucret:.2f} TL")
print(f"KDV (%20) : {test_ucret * 0.20:.2f} TL")
print(f"KDV Dahil : {sonuc:.2f} TL")
```

```python
# ========================
# GÖREV 3: Kütüphane Kullanımı — math ve datetime
# ========================

# Tüm müşterilerin KDV'li fatura tutarlarını hesaplama — For döngüsü
print("=== FATURA RAPORU ===")
toplam_tutar = 0.0                               # Float — toplam tutar sayacı

for musteri in musteriler_listesi:               # Döngü Kullanımı — liste üzerinde gezinme
    kdv_dahil = tutar_hesapla(musteri["aylik_ucret"])   # Fonksiyon çağrısı
    yuvarlanmis = math.ceil(kdv_dahil)                  # math kütüphanesi — yukarı yuvarlama
    toplam_tutar += kdv_dahil                           # Birikimli toplama
    print(f"  {musteri['ad_soyad']:20} → KDV Dahil: {kdv_dahil:>8.2f} TL  |  Yuvarlak: {yuvarlanmis} TL")

# math kütüphanesi ile toplam fatura tutarını yukarı yuvarlama
toplam_yuvarlanmis = math.ceil(toplam_tutar)     # math.ceil() — yukarı yuvarlama
print(f"\n  {'TOPLAM FATURA':20} → KDV Dahil: {toplam_tutar:>8.2f} TL  |  Yuvarlak: {toplam_yuvarlanmis} TL")

# datetime kütüphanesi ile bugünün tarihini faturanın altına ekleme
bugun = datetime.datetime.now()                  # datetime kütüphanesi — anlık tarih/saat
tarih_format = bugun.strftime("%d/%m/%Y %H:%M") # strftime — tarih biçimlendirme
print(f"\n  📅 Fatura Tarihi: {tarih_format}")
```

```python
# ========================
# GÖREV 4: Hata Denetimi ve Set Kullanımı
# ========================

# Tüm hizmetleri tekrarlı olarak bir listeye atma
tum_hizmetler_tekrarli = []                           # Boş liste oluşturma

for musteri in musteriler_listesi:                    # Döngü Kullanımı — müşteriler üzerinde
    for hizmet in musteri["hizmetler"]:               # İç içe döngü — hizmetler üzerinde
        tum_hizmetler_tekrarli.append(hizmet)         # Liste metodu — eleman ekleme

print("=== TÜM HİZMETLER (Tekrarlı) ===")
print(f"  Liste: {tum_hizmetler_tekrarli}")
print(f"  Eleman sayısı: {len(tum_hizmetler_tekrarli)}")

# Set() kullanımı — Tekrar eden hizmetleri temizleme
benzersiz_hizmetler = set(tum_hizmetler_tekrarli)     # Set Kullanımı — benzersiz elemanlar

print("\n=== BENZERSİZ HİZMETLER (Set ile) ===")
print(f"  Set: {benzersiz_hizmetler}")
print(f"  Benzersiz hizmet sayısı: {len(benzersiz_hizmetler)}")

# Hata Denetimi — try/except ile güvenli veri erişimi
print("\n=== HATA DENETİMİ (Try-Except) ===")
try:                                                  # Try bloğu — hata çıkabilecek işlem
    test_deger = int("abc")                           # Kasıtlı hata: string'i int'e çevirme
except ValueError as hata:                           # Except bloğu — hatayı yakala
    print(f"  ⚠️  Hata Yakalandı: {hata}")           # Hata mesajı yazdırma
    print(f"  ✅ Program çalışmaya devam ediyor.")

# Gerçek kullanım: sözlük erişiminde hata denetimi
try:
    deger = musteriler_listesi[0]["var_olmayan_alan"]  # KeyError üretecek
except KeyError as hata:                              # KeyError — olmayan anahtar
    print(f"  ⚠️  Sözlük Hatası: {hata} anahtarı bulunamadı.")
```

```python
# ========================
# GÖREV 5: AI Denetimi + Churn (Müşteri Kaybı) Analizi
# (Bu kod önce AI'ya yazdırıldı, hatalar düzeltildi ve satır yorumları eklendi)
# ========================

def churn_analizi(musteriler):                        # Fonksiyon Tanımlama — churn tespiti
    """
    Müşteri listesini döngüyle gezer, churn riskini analiz eder.
    Kriter 1: Sadakat ayı 6'dan az  → Yeni müşteri, henüz bağlı değil
    Kriter 2: aktif_mi = False       → Hizmet kullanmıyor, ayrılma sinyali
    Kriter 3: Aylık ücret < 200 TL  → Düşük değerli müşteri, kolayca vazgeçer
    """
    churn_riskli = []                                 # Liste — churn riskli müşteriler
    guvenli = []                                      # Liste — güvenli müşteriler

    for musteri in musteriler:                        # Döngü Kullanımı — listeyi gezme
        risk_puani = 0                                # Integer — risk puanı sayacı
        risk_sebebi = []                              # Liste — risk sebepleri

        # Kriter 1: Sadakat süresi kontrolü
        if musteri["sadakat_ayi"] < 6:                # Koşullu İfade
            risk_puani += 2                           # Aritmetik — puan artırma
            risk_sebebi.append("Sadakat süresi düşük (<6 ay)")  # Liste metodü

        # Kriter 2: Aktiflik durumu kontrolü
        if not musteri["aktif_mi"]:                   # Boolean kontrolü
            risk_puani += 3
            risk_sebebi.append("Hesap pasif")

        # Kriter 3: Düşük ücret kontrolü
        if musteri["aylik_ucret"] < 200:              # Float karşılaştırma
            risk_puani += 1
            risk_sebebi.append("Düşük aylık ücret (<200 TL)")

        # Risk puanına göre sınıflandırma
        if risk_puani >= 2:                           # Koşullu İfade — risk eşiği
            churn_riskli.append({
                "ad": musteri["ad_soyad"],            # String — müşteri adı
                "puan": risk_puani,                   # Integer — risk puanı
                "sebepler": risk_sebebi               # Liste — sebep listesi
            })
        else:
            guvenli.append(musteri["ad_soyad"])       # Güvenli müşteri listesine ekle

    return churn_riskli, guvenli                      # Return — iki değer döndürme (Tuple)


# Fonksiyonu çalıştırma
churn_riskli, guvenli_musteriler = churn_analizi(musteriler_listesi)  # Tuple unpacking

print("=" * 50)
print("       CHURN (AYRILMA RİSKİ) ANALİZİ")
print("=" * 50)

print(f"\n🚨 Churn Riskli Müşteriler ({len(churn_riskli)} kişi):")
for m in churn_riskli:                               # Döngü Kullanımı — riskli müşteriler
    print(f"  ❌ {m['ad']:20} | Risk Puanı: {m['puan']} | Sebepler: {', '.join(m['sebepler'])}")

print(f"\n✅ Güvenli Müşteriler ({len(guvenli_musteriler)} kişi):")
for ad in guvenli_musteriler:                        # Döngü Kullanımı — güvenli müşteriler
    print(f"  ✓ {ad}")

print(f"\n📊 Churn Oranı: %{(len(churn_riskli)/len(musteriler_listesi)*100):.0f}")
```

### 💬 Kritik Soru Cevabı — 2. Kısım

> **"Yazdığınız döngü (for loop) müşteri listesini gezerken, hangi müşterilerin Churn (ayrılma) riskinde olduğunu nasıl tespit ettiniz? Mantığınızı açıklayın."**

**Cevap:**

Churn tespiti için **ağırlıklı risk puanı** mantığı kullandık. `for` döngüsü her müşteride sırayla şu üç kriteri kontrol eder:

| Kriter | Koşul | Risk Puanı |
|--------|-------|------------|
| Sadakat süresi | `sadakat_ayi < 6` | +2 puan |
| Hesap durumu | `aktif_mi == False` | +3 puan |
| Aylık ücret | `aylik_ucret < 200 TL` | +1 puan |

Her müşterinin risk puanı toplanır. **2 veya üzeri puan** alan müşteriler "Churn Riskli" olarak sınıflandırılır.

**Gerekçe:**
- Hesabı pasif olan müşteri zaten hizmeti kullanmıyor → ayrılma olasılığı yüksek (ağır puan: 3).
- Sadakat süresi çok kısa olan müşteri henüz bağ kurmamış → kolayca rakibe geçebilir (orta puan: 2).
- Düşük ücretli müşteri az hizmet kullanıyor → ek bir sorunla birlikte ayrılabilir (hafif puan: 1).

Bu yöntemle yalnızca tek kriterle değil, **birden fazla kriterin bir araya gelmesiyle** daha isabetli bir churn tespiti yapılmıştır.
